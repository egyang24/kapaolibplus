## 0 scalar gain, 0.8 tip/tilt¶

##### This set of notebooks examines 4 sets of open loop data with a slow moving atmosphere taken in the lab on 6/20/14: 0 scalar gain 0 tip/tilt, .8 scalar gain 0 tip/tilt, 0 scalar gain .8 tip/tilt, .8 both. This specific notebook looks at the case in which we have 0 scalar gain, 0.8 tip/tilt¶

```python
%pylab inline
```

```python
from kapaolibplus import (subaps_to_grid, overlay_indices, overlay_subap_values, parse_telem_table, 
    Slopes, IntensityMap, DMPositions, load_reconstructor, newpos_to_grid, overlay_indices_newpos)
```

```python
zero_sg = '20140621_000310'
SLOPE_X_FILE = './TelemetryOpenLoop/slope_x_' + zero_sg + '.tel'
SLOPE_Y_FILE = './TelemetryOpenLoop/slope_y_' + zero_sg + '.tel'
INTENSITY_MAP = './TelemetryOpenLoop/intensity_map_' + zero_sg + '.tel'
NEWPOS_FILE = './TelemetryOpenLoop/new_pos_' + zero_sg + '.tel'
print SLOPE_X_FILE
print SLOPE_Y_FILE
print INTENSITY_MAP
print NEWPOS_FILE
```

```python
slope_x, slope_y, intensity_map = Slopes(SLOPE_X_FILE), Slopes(SLOPE_Y_FILE), IntensityMap(INTENSITY_MAP)
```

```python
slope_x.new_overall = np.median(slope_x.data[:,intensity_map.include_mask], axis=1)
slope_y.new_overall = np.median(slope_y.data[:,intensity_map.include_mask], axis=1)

figure(figsize=(10,5))
plot(slope_x.overall_slopes, label="Old Overall")
plot(slope_x.new_overall, label="Masked Overall")
xlabel("Time step")
ylabel("X slope")
legend()

figure(figsize=(10,5))
plot(slope_y.overall_slopes, label="Old Overall")
plot(slope_y.new_overall, label="Masked Overall")
xlabel("Time step")
ylabel("Y slope")
legend()
```

### Compare raw data to data with overall signal removed¶

#### Raw data for good subaps (colored lines) with new_overall timeseries overplotted in white

```python
xmax = np.max(slope_y.timestamps - slope_y.timestamps[0]) / np.timedelta64(1,'ms')

figure(figsize=(6,3))
ax = subplot(211)
_ = plot(slope_x.timestamps - slope_x.timestamps[0], slope_x.data[:,intensity_map.include_mask], alpha=0.1)
_ = plot(slope_x.timestamps - slope_x.timestamps[0], slope_x.new_overall, c='white')
ylim(-0.4, 0.4)
yticks([-0.4, -0.2, 0, 0.2, 0.4])
locs, _ = xticks()
xticks(locs, [])
xlim(0, xmax)
ylabel("X Slope")

subplot(212)
_ = plot(slope_y.timestamps - slope_y.timestamps[0], slope_y.data[:,intensity_map.include_mask], alpha=0.1)
_ = plot(slope_y.timestamps - slope_y.timestamps[0], slope_y.new_overall, c='white')
ylim(-0.4, 0.4)
yticks([-0.4, -0.2, 0, 0.2, 0.4])
ylabel("Y Slope")
xlabel("Time (ms)")
xlim(0, xmax)
```

#### Good subaperture slopes after subtracting off overall trend (colored lines)

```python
xmax = np.max(slope_y.timestamps - slope_y.timestamps[0]) / np.timedelta64(1,'ms')

figure(figsize=(6,3))
ax = subplot(211)
_ = plot(slope_x.timestamps - slope_x.timestamps[0], slope_x.data[:,intensity_map.include_mask] - slope_x.new_overall[:, np.newaxis], alpha=0.1)
axhline(0, color='white')
ylim(-0.4, 0.4)
yticks([-0.4, -0.2, 0, 0.2, 0.4])
locs, _ = xticks()
xticks(locs, [])
xlim(0, xmax)
ylabel("X Slope")

subplot(212)
_ = plot(slope_y.timestamps - slope_y.timestamps[0], slope_y.data[:,intensity_map.include_mask] - slope_y.new_overall[:, np.newaxis], alpha=0.1)
axhline(0, color='white')
ylim(-0.4, 0.4)
yticks([-0.4, -0.2, 0, 0.2, 0.4])
ylabel("Y Slope")
xlabel("Time (ms)")
xlim(0, xmax)
```

```python
# in the next cell we're plotting power in 1 subap... but what's a typical subap? Let's pick one from the intensity map with a high median value

imshow(subaps_to_grid(intensity_map.median_by_subap), origin='lower')
overlay_indices()
colorbar()
```

### Looking at power spectra

```python
figure(figsize=(6,5))

subplot(211)
# subap 46 is chosen as a 'typical' subap for its good median value... check it out with the intensity map and subaps_to_grid
_ = plot(slope_x.freqs, slope_x.power[:,46], label="Power in Subap 46 (X)") 

# here we're FFTing the new_overall time series, taking magnitude w/ np.abs
plot(slope_x.freqs, np.abs(np.fft.fft(slope_x.new_overall)), label="Power in Overall (Median) Slope")

# for reference, what does an f^{-2/3} slope look like? better plot it
plot(slope_x.freqs, slope_x.freqs**(-2.0/3.0) * 10.0, label="$f\,^{-2/3}$") 

# set limits based on fft freqs (only positive, b/c loglog plot)
xlim(np.min(slope_x.freqs[slope_x.freqs > 0]), np.max(slope_x.freqs))
ylim(1, 500)
xscale('log', nonposy='mask') # mask out non positive values in loglog plot
yscale('log', nonposy='mask')
ylabel('Power (slope data units)')
legend(fontsize=10)

# then the same for Y
subplot(212)
_ = plot(slope_y.freqs, slope_y.power[:,46], label="Power in Subap 46 (Y)")
plot(slope_y.freqs, np.abs(np.fft.fft(slope_y.new_overall)), label="Power in Overall (Median) Slope")
plot(slope_y.freqs, slope_y.freqs**(-2.0/3.0) * 10.0, label="$f\,^{-2/3}$") 
xlim(np.min(slope_y.freqs[slope_y.freqs > 0]), np.max(slope_y.freqs))
ylim(1, 500)
xscale('log', nonposy='mask')
yscale('log', nonposy='mask')
xlabel('Frequency (Hz)')
ylabel('Power (slope data units)')
legend(fontsize=10)

tight_layout()
```

### Compute the FFT with the better-overall-subtracted data¶

```python
subtracted_x_data = slope_x.data - slope_x.new_overall[:,np.newaxis]
x_freqs, x_power, x_subtracted_power = slope_x.do_fft(subtracted_data=subtracted_x_data)
```

```python
subtracted_y_data = slope_y.data - slope_y.new_overall[:,np.newaxis]
y_freqs, y_power, y_subtracted_power = slope_y.do_fft(subtracted_data=subtracted_y_data)
```

```python
figure(figsize=(6,5))

subplot(211)
plot(x_freqs, np.average(slope_x.power[:,intensity_map.include_mask], axis=1), label="Average Power, Good Subaps (X)")
plot(x_freqs, np.average(x_subtracted_power[:,intensity_map.include_mask], axis=1), label="Residual Power, Good Subaps (X)")
plot(x_freqs, x_freqs**(-2.0/3.0) * 10.0, label="$f\,^{-2/3}$")
plot(x_freqs, x_freqs**(4.0/3.0) * 100.0, label="$f\,^{4/3}$")
xlim(np.min(x_freqs[x_freqs > 0]), np.max(x_freqs))
ylim(1, 500)
xscale('log', nonposy='mask')
yscale('log', nonposy='mask')
ylabel('Power (slope data units)')
legend(fontsize=10)

subplot(212)
plot(y_freqs, np.average(slope_y.power[:,intensity_map.include_mask], axis=1), label="Average Power, Good Subaps (Y)")
plot(y_freqs, np.average(y_subtracted_power[:,intensity_map.include_mask], axis=1), label="Residual Power, Good Subaps (Y)")
plot(y_freqs, y_freqs**(-2.0/3.0) * 10.0, label="$f\,^{-2/3}$")
plot(y_freqs, y_freqs**(4.0/3.0) * 100.0, label="$f\,^{4/3}$")
xlim(np.min(y_freqs[y_freqs > 0]), np.max(y_freqs))
ylim(1, 500)
xscale('log', nonposy='mask')
yscale('log', nonposy='mask')
xlabel('Frequency (Hz)')
ylabel('Power (slope data units)')
legend(fontsize=10)

tight_layout()
```

#### Tip/tilt as a function of time

```python
#pulling out the 120th and 122nd index for each time step
length = len(new_pos.data)
tt_1 = zeros(length)
tt_2 = zeros(length)
for i in range(0,length):
    tt_1[i] = new_pos.data[i][120]
    tt_2[i] = new_pos.data[i][122]
```

```python
subplot(2,1,1)
plot(new_pos.timestamps - new_pos.timestamps[0],tt_1)
ylabel("Tip/tilt index 120")

subplot(2,1,2)
plot(new_pos.timestamps - new_pos.timestamps[0],tt_2)
ylabel("Tip/tilt index 122")
xlabel("Time (ms)")
```
