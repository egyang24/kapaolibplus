## Analysis Template, Consolidated

##### This notebook takes a set of telemetry files and analyzes/visualizes the slopes, intensities, and actuator positions.

```python
%pylab inline
from IPython.html.widgets import *
from kapaolibplus import (subaps_to_grid, overlay_indices, overlay_subap_values, parse_telem_table, 
    Slopes, IntensityMap, DMPositions, load_reconstructor, newpos_to_grid, overlay_indices_newpos, slope_to_grid, overlay_indices_slope)
```

#### Change the file number and/or directory here:

```python
filenum = '20140710_214914'
SLOPE_X_FILE = 'slope_x_' + filenum + '.tel'
SLOPE_Y_FILE = 'slope_y_' + filenum + '.tel'
INTENSITY_MAP = 'intensity_map_' + filenum + '.tel'
NEWPOS_FILE = 'new_pos_' + filenum + '.tel'
```

#### Slopes and intensity maps

```python
slope_x, slope_y, intensity_map = Slopes(SLOPE_X_FILE), Slopes(SLOPE_Y_FILE), IntensityMap(INTENSITY_MAP)
new_pos = DMPositions(NEWPOS_FILE)



length = len(slope_x.data)

#finding min and max values at each of the 3 relevant time steps, then the overall min and max values between the 3

# X slope
xmin_t0 = min(slope_x.data[0])
xmin_mid = min(slope_x.data[int(length/2)])
xmin_last = min(slope_x.data[length - 1])
xmax_t0 = max(slope_x.data[0])
xmax_mid = max(slope_x.data[int(length/2)])
xmax_last = max(slope_x.data[length - 1])

xmin = min(xmin_t0,xmin_mid,xmin_last)
xmax = max(xmax_t0,xmax_mid,xmax_last)


# Y slope

ymin_t0 = min(slope_y.data[0])
ymin_mid = min(slope_y.data[int(length/2)])
ymin_last = min(slope_y.data[length - 1])
ymax_t0 = max(slope_y.data[0])
ymax_mid = max(slope_y.data[int(length/2)])
ymax_last = max(slope_y.data[length - 1])

ymin = min(ymin_t0,ymin_mid,ymin_last)
ymax = max(ymax_t0,ymax_mid,ymax_last)


# Intensities

intensmin_t0 = min(intensity_map.data[0])
intensmin_mid = min(intensity_map.data[int(length/2)])
intensmin_last = min(intensity_map.data[length - 1])
intensmax_t0 = max(intensity_map.data[0])
intensmax_mid = max(intensity_map.data[int(length/2)])
intensmax_last = max(intensity_map.data[length - 1])

intensmin = min(intensmin_t0,intensmin_mid,intensmin_last)
intensmax = max(intensmax_t0,intensmax_mid,intensmax_last)


# Newpos

newposshort_t0 = zeros(120)
newposshort_mid = zeros(120)
newposshort_last = zeros(120)

for i in range(120):
    newposshort_t0[i] = new_pos.data[0][i]
    newposshort_mid[i] = new_pos.data[int(length/2)][i]
    newposshort_last[i] = new_pos.data[length - 1][i]
    
newposmin_t0 = min(newposshort_t0)
newposmin_mid = min(newposshort_mid)
newposmin_last = min(newposshort_last)
newposmax_t0 = max(newposshort_t0)
newposmax_mid = max(newposshort_mid)
newposmax_last = max(newposshort_last)

newposmin = min(newposmin_t0,newposmin_mid,newposmin_last)
newposmax = max(newposmax_t0,newposmax_mid,newposmax_last)



# Plots


def xslopemapfixed(time):
    imshow(slope_to_grid(slope_x.data[time]), origin='lower',vmin = xmin, vmax = xmax)
    overlay_indices_slope()
    colorbar()
    
figure(figsize=(20,5))
subplot(141)
xslopemapfixed(0)
title('first time step')

subplot(142)
xslopemapfixed(int(length/2))
title('middle time step')

subplot(143)
xslopemapfixed(length-1)
title('last time step')

subplot(144)
imshow(slope_to_grid(slope_x.median_by_slope), origin='lower',vmin = xmin, vmax = xmax)
overlay_indices()
colorbar()
title('median slopes')

suptitle('X Slope Maps, Fixed',fontsize=16)


def xslopemap(time):
    imshow(slope_to_grid(slope_x.data[time]), origin='lower')
    overlay_indices_slope()
    colorbar()
    
figure(figsize=(20,5))
subplot(141)
xslopemap(0)
title('first time step')

subplot(142)
xslopemap(int(length/2))
title('middle time step')

subplot(143)
xslopemap(length-1)
title('last time step')

subplot(144)
imshow(slope_to_grid(slope_x.median_by_slope), origin='lower')
overlay_indices()
colorbar()
title('median slopes')

suptitle('X Slope Maps, Unfixed',fontsize=16)






def yslopemapfixed(time):
    imshow(slope_to_grid(slope_y.data[time]), origin='lower',vmin = ymin, vmax = ymax)
    overlay_indices_slope()
    colorbar()
    
figure(figsize=(20,5))
subplot(141)
yslopemapfixed(0)
title('first time step')

subplot(142)
yslopemapfixed(int(length/2))
title('middle time step')

subplot(143)
yslopemapfixed(length-1)
title('last time step')

subplot(144)
imshow(slope_to_grid(slope_y.median_by_slope), origin='lower',vmin = ymin, vmax = ymax)
overlay_indices()
colorbar()
title('median slopes')

suptitle('Y Slope Maps, Fixed',fontsize=16)



def yslopemap(time):
    imshow(slope_to_grid(slope_y.data[time]), origin='lower')
    overlay_indices_slope()
    colorbar()
    
figure(figsize=(20,5))
subplot(141)
yslopemap(0)
title('first time step')

subplot(142)
yslopemap(int(length/2))
title('middle time step')

subplot(143)
yslopemap(length-1)
title('last time step')

subplot(144)
imshow(slope_to_grid(slope_y.median_by_slope), origin='lower')
overlay_indices()
colorbar()
title('median slopes')

suptitle('Y Slope Maps, Unfixed',fontsize=16)





def intensmapfixed(time):
    imshow(subaps_to_grid(intensity_map.data[time]), origin='lower',vmin = intensmin, vmax = intensmax)
    overlay_indices()
    colorbar()
    
figure(figsize=(20,5))
subplot(141)
intensmapfixed(0)
title('first time step')

subplot(142)
intensmapfixed(int(length/2))
title('middle time step')

subplot(143)
intensmapfixed(length-1)
title('last time step')

subplot(144)
imshow(subaps_to_grid(intensity_map.median_by_subap), origin='lower',vmin = intensmin, vmax = intensmax)
overlay_indices()
colorbar()
title('median intensity')

suptitle('Intensity Maps, Fixed',fontsize=16)


def intensmap(time):
    imshow(subaps_to_grid(intensity_map.data[time]), origin='lower')
    overlay_indices()
    colorbar()
    
figure(figsize=(20,5))
subplot(141)
intensmap(0)
title('first time step')

subplot(142)
intensmap(int(length/2))
title('middle time step')

subplot(143)
intensmap(length-1)
title('last time step')

subplot(144)
imshow(subaps_to_grid(intensity_map.median_by_subap), origin='lower')
overlay_indices()
colorbar()
title('median intensity')

suptitle('Intensity Maps, Unfixed',fontsize=16)
```

#### Newpos analysis: new pos maps and tip/tilt

```python
def posmapfixed(time):
    imshow(newpos_to_grid(new_pos.data[time]), origin='lower',vmin = newposmin, vmax = newposmax)
    overlay_indices_newpos()
    colorbar()
    
figure(figsize=(20,5))
subplot(141)
posmapfixed(0)
title('first time step')

subplot(142)
posmapfixed(int(length/2))
title('middle time step')

subplot(143)
posmapfixed(length-1)
title('last time step')

subplot(144)
imshow(newpos_to_grid(new_pos.median_by_pos), origin='lower',vmin = newposmin, vmax = newposmax)
overlay_indices_newpos()
colorbar()
title('median positions')

suptitle('DM Position (New_Pos) Maps, Fixed',fontsize=16)



def posmap(time):
    imshow(newpos_to_grid(new_pos.data[time]), origin='lower')
    overlay_indices_newpos()
    colorbar()
    
figure(figsize=(20,5))
subplot(141)
posmap(0)
title('first time step')

subplot(142)
posmap(int(length/2))
title('middle time step')

subplot(143)
posmap(length-1)
title('last time step')

subplot(144)
imshow(newpos_to_grid(new_pos.median_by_pos), origin='lower')
overlay_indices_newpos()
colorbar()
title('median positions')

suptitle('DM Position (New_Pos) Maps, Unfixed',fontsize=16)



#pulling out the 120th and 122nd index for each time step
length = len(new_pos.data)
tt_1 = zeros(length)
tt_2 = zeros(length)
for i in range(0,length):
    tt_1[i] = new_pos.data[i][120]
    tt_2[i] = new_pos.data[i][122]
    
figure(figsize=(12,6))
subplot(2,1,1)
plot(new_pos.timestamps - new_pos.timestamps[0],tt_1)
ylabel("Tip/tilt index 120")

subplot(2,1,2)
plot(new_pos.timestamps - new_pos.timestamps[0],tt_2)
ylabel("Tip/tilt index 122")
xlabel("Time (ms)")

suptitle('Tip/tilt as a function of time',fontsize=16)
```
